# Plotting the retweets on graphs

```python
def round_5(x, base=5):
    return int(base * math.ceil(x/base))
```

```python
def check_same_period(time1, time2, period):
    if time2 > time1:
        if (time2 - time1).seconds <= period:
            #print((time2 - time1).seconds)
            return True
        else:
            return False
    else:
        if (time1 - time2).seconds <= period:
            return True
        else:
            return False
```

```python
def plot_retweets_main(account):
    tweets_list = top_prolific_ira_retweets[account].keys()
    all_retweets = []

    for tweet in tweets_list:
        if len(top_prolific_ira_retweets[account][tweet]['retweets']) > 0: 
            # data structures to hold data
            retweets_list = []
            retweets_freq_list = []

            # convert the original retweet date to date time
            original_tweet_asdatetime = dateutil.parser.parse(top_prolific_ira_retweets[account][tweet]['original_date'])
            
            # convert the retweets' dates to date time
            for retweet_time in top_prolific_ira_retweets[account][tweet]['retweets']:
                retweet_datetime = dateutil.parser.parse(retweet_time)
                
                checked = False
                for compared_time in retweets_list:
                    if check_same_period(compared_time, retweet_datetime, 3600):
                        retweets_freq_list[retweets_list.index(compared_time)] = retweets_freq_list[retweets_list.index(compared_time)] + 1
                        checked = True
                        break
                if checked:
                    checked = False
                else:
                    retweets_list.append(retweet_datetime)
                    retweets_freq_list.append(1)
            all_retweets.append([original_tweet_asdatetime, retweets_list, retweets_freq_list, tweet])
    
    if len(all_retweets) < 1:
        return
    else:
        max_num = max(max(map(lambda x: x[2], all_retweets)))
        #print(max_num)
        plot_retweets(all_retweets, account)
    
```

```python
def plot_retweets(all_retweets, account):
     
    bottom = dates.date2num(dateutil.parser.parse('2016-10-20 00:00:00+00:00'))
    top = dates.date2num(dateutil.parser.parse('2016-11-01 23:59:00+00:00'))
    
    num_plots = len(all_retweets)
    
    if num_plots <= 1:
        
        ## size
        figsize = (12, 3)
        plt.figure(figsize = figsize) # size of figure

        ## colors
        rand_color = randomcolor.RandomColor()
        plot_color = rand_color.generate()
        
        ## axis labels, ticks
        plt.ylabel('Number of Retweets',  fontsize=12) 
        plt.xlabel('Date',  fontsize=12)
        
        plt.xlim((bottom, top)) # setting the range(by date) on the x-axis
        plt.yticks(np.arange(max(all_retweets[0][2])+ 1)) # setting the y-ticks
        
        ## title and text
        plt.title('Retweet Upticks: @%s' % account, fontsize=14, fontweight='bold', y=1.05)
        
        ## actual plotting
        plt.plot_date(dates.date2num(all_retweets[0][1]), all_retweets[0][2], 'o', color = plot_color[0])
        plt.axvline(x=dates.date2num(all_retweets[0][0]), color= plot_color[0])

    else:
        ''' for retweet in all_retweets:
            print(retweet[2])'''
    
        ## size, titles, location
        figsize = (12, 3*num_plots)
        fig, axs = plt.subplots(num_plots, figsize=figsize, sharex=True, sharey = True)
        fig.suptitle('Retweet Upticks: @%s' % account, fontsize=14, fontweight='bold', y=1.05)
        fig.tight_layout(pad=3.5)
        
        # labels and ticks
        plt.ylabel('Number of Retweets',  fontsize=12) 
        plt.xlabel('Date',  fontsize=12) 
        plt.xlim((bottom, top))
        
        max_num = max(max(map(lambda x: x[2], all_retweets))) +1
        max_ytick = round_5(max_num)
        #print(max_ytick)
        
        # actual plotting
        for i in range(0, num_plots):
            rand_color = randomcolor.RandomColor()
            plot_color = rand_color.generate()
            
            #arr = np.arange(0, max_ytick + 1, step = max_ytick/5)
            #ytickslocs = arr
            #print(arr)
            
            #axs[i].yaxis.set_major_locator(mticker.FixedLocator(ytickslocs))
            #axs[i].yticks(np.arange(0, 81, 10))
            #axs[i].set_yticks(np.arange(0, 6, 1))
            
            axs[i].set_title('original tweet:\n%s' % all_retweets[i][3])
            axs[i].plot_date(dates.date2num(all_retweets[i][1]), all_retweets[i][2], 'o', color = plot_color[0])
            axs[i].axvline(x=dates.date2num(all_retweets[i][0]), color= plot_color[0])
    plt.savefig('%sretweets.pdf'% (account))
    plt.show()
```

# LDA Model for Retweeted vs Not Retweeted Tweets

```python
count_vectorizer = CountVectorizer(stop_words='english')
```

```python
def plot_10_most_common_words(count_data, count_vectorizer, title):
    import matplotlib.pyplot as plt
    words = count_vectorizer.get_feature_names()
    total_counts = np.zeros(len(words))
    for t in count_data:
        total_counts+=t.toarray()[0]
    
    count_dict = (zip(words, total_counts))
    count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:10]
    words = [w[0] for w in count_dict]
    counts = [w[1] for w in count_dict]
    x_pos = np.arange(len(words)) 
    
    plt.figure(2, figsize=(15, 15/1.6180))
    plt.subplot(title='10 most common words %s' % title)
    sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.5})
    sns.barplot(x_pos, counts, palette='husl')
    plt.xticks(x_pos, words, rotation = 45, ha = 'right') 
    plt.xlabel('words')
    plt.ylabel('counts')
    plt.savefig('plot.pdf')  
    plt.show()
```

```python
tweets_retweeted_df['text_processed'] = tweets_retweeted_df['text'].apply(lambda x: re.sub(r"http\S+", "", str(x)))
tweets_retweeted_df['text_processed'] = tweets_retweeted_df['text_processed'].apply(lambda x: ' '.join(word.strip(string.punctuation) for word in x.split()))
tweets_retweeted_df['text_processed'] = tweets_retweeted_df['text_processed'].apply(lambda x: x.lower())
```

```python
count_data_notretweeted = count_vectorizer.fit_transform(tweets_retweeted_df[tweets_retweeted_df['retweeted'] == "Not Retweeted"]['text_processed'])
```

```python
plot_10_most_common_words(count_data_notretweeted, count_vectorizer, "for tweets not retweeted")
```

```python
count_data_retweeted = count_vectorizer.fit_transform(tweets_retweeted_df[tweets_retweeted_df['retweeted'] == "Retweeted"]['text_processed'])
```

```python
plot_10_most_common_words(count_data_retweeted, count_vectorizer, "for tweets retweeted")
```

```python
import warnings
warnings.simplefilter("ignore", DeprecationWarning)
# Load the LDA model from sk-learn
from sklearn.decomposition import LatentDirichletAllocation as LDA
 
# Helper function
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
# Tweak the two parameters below
number_topics = 5
number_words = 10
# Create and fit the LDA model
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data_retweeted)
# Print the topics found by the LDA model
print("Topics found for tweets retweeted via LDA:")
print_topics(lda, count_vectorizer, number_words)
```

```python
lda = LDA(n_components=number_topics, n_jobs=-1)

lda.fit(count_data_notretweeted)
# Print the topics found by the LDA model
print("Topics for tweets not retweeted found via LDA:")
print_topics(lda, count_vectorizer, number_words)
```

# Getting a list of all the retweeters and the tweets that they retweeted
#### key = retweeter account's handle

```python
all_retweeters = {}
for account in all_retweeted_accounts.keys():
    for retweet in all_retweeted_accounts[account]:
        for i in range(0, len(all_retweeted_accounts[account][retweet]['retweeters'])):
            retweeter_handle = all_retweeted_accounts[account][retweet]['retweeters'][i][0]
            retweeter_numfollowers = all_retweeted_accounts[account][retweet]['retweeters'][i][1]
            retweeter_bio = all_retweeted_accounts[account][retweet]['retweeters'][i][2]
            retweet_date = all_retweeted_accounts[account][retweet]['retweets'][i]
            if all_retweeters.get(retweeter_handle) == None:
                all_retweeters[retweeter_handle] = {'user_bio': retweeter_bio, "num_followers": retweeter_numfollowers, account:[retweet_date]}
            else:
                if all_retweeters[retweeter_handle].get(account) != None:
                    all_retweeters[retweeter_handle][account].append(retweet_date)
                else:
                    all_retweeters[retweeter_handle][account] = [retweet_date]
                
```
