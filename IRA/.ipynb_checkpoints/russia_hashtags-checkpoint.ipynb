```python
with open("russia.csv") as csv_file:
    russia_all_csv = pd.read_csv(csv_file)
    
```

```python
russia_all_csv.head()
```

```python
stored_list = {}
for index, row in russia_all_csv.iterrows():
    tweet = str(row['content']).split()
    for word in tweet:
        if word[0] == "#":
            if stored_list.get(word) == None:
                stored_list[word] = 1
            else:
                stored_list[word] += 1
```

```python
with open('test.csv', 'w') as f:
    for key in stored_list.keys():
        new_key = key.replace(',', '');
        f.write("%s,%s\n"%(new_key,stored_list[key]))
```
